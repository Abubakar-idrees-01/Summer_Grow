## Orthogonality, Projections, Norms, Distance

We continue using the same House Price dataset.

```python
import numpy as np
import pandas as pd

np.set_printoptions(suppress=True)

data = {
    'Area': [1000, 1500, 2000, 2500, 3000],
    'Bedrooms': [2, 3, 3, 4, 4],
    'Age': [10, 5, 20, 15, 8],
    'Distance': [5, 3, 10, 2, 7],
    'Price': [200000, 300000, 250000, 400000, 500000]
}

df = pd.DataFrame(data)

X = df[['Area','Bedrooms','Age','Distance']].to_numpy()
X
```

# 1. Orthogonality

Two vectors are orthogonal if their dot product is zero.

Meaning:
- No relationship
- Independent directions

In ML:
- Orthogonal features = no redundancy

```python
a = np.array([1, 0])
b = np.array([0, 1])

np.dot(a, b)
```

# 2. Norms (Vector Length)

Norm = size or magnitude of a vector.

Used in ML for:
- Regularization
- Error measurement
- Distance calculation

```python
v = np.array([3, 4])

np.linalg.norm(v)
```

# 3. Types of Norms

- L1 Norm → sum of absolute values
- L2 Norm → Euclidean distance
- L∞ Norm → max value

Used in:
- Lasso (L1)
- Ridge (L2)

```python
v = np.array([3, -4, 2])

l1 = np.sum(np.abs(v))
l2 = np.linalg.norm(v)
linf = np.max(np.abs(v))

l1, l2, linf
```

# 4. Distance Between Points

Used to measure similarity between data points.

```python
p1 = np.array([1000, 2])
p2 = np.array([1500, 3])

np.linalg.norm(p1 - p2)
```

# 5. Cosine Similarity

Measures direction similarity, not magnitude.

Used in:
- Recommendation systems
- NLP embeddings
- Search engines

```python
a = np.array([1,2,3])
b = np.array([2,4,6])

cos_sim = np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b))
cos_sim
```

# 6. Projection (Very Important)

Projection = shadow of one vector on another.

Used in:
- Linear Regression
- PCA
- Dimensionality reduction

```python
a = np.array([3, 4])
b = np.array([5, 0])

proj = (np.dot(a, b) / np.dot(b, b)) * b
proj
```

# 7. Why this matters in Data Science

- KNN uses distance
- Recommendation systems use cosine similarity
- PCA uses projections
- Regression minimizes norm of error
- Embeddings use vector similarity

# Key Takeaways

- Orthogonality = independence
- Norm = size of vector
- Distance = similarity measure
- Cosine similarity = direction-based similarity
- Projection = foundation of PCA & regression
