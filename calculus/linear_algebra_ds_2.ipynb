# Linear Algebra for Data Science - Part 2
## Matrix Multiplication, Transpose, Inverse, Determinants

We continue using the same House Price dataset.

```python
import numpy as np
import pandas as pd

np.set_printoptions(suppress=True)

data = {
    'Area': [1000, 1500, 2000, 2500, 3000],
    'Bedrooms': [2, 3, 3, 4, 4],
    'Age': [10, 5, 20, 15, 8],
    'Distance': [5, 3, 10, 2, 7],
    'Price': [200000, 300000, 250000, 400000, 500000]
}

df = pd.DataFrame(data)

X = df[['Area','Bedrooms','Age','Distance']].to_numpy()
y = df['Price'].to_numpy()

X, y
```

# 1. Matrix Multiplication

Used to compute predictions in Machine Learning models.

```python
w = np.array([120, 50000, -3000, -8000])

pred = X @ w
pred
```

# 2. Transpose

Flips rows and columns. Used in ML formulas.

```python
X.T
```

# 3. Matrix Inverse

Used to solve linear equations (Linear Regression).

```python
A = np.array([
    [2,1],
    [1,3]
])

np.linalg.inv(A)
```

Important: In real ML, inverse is rarely used directly due to instability.

# 4. Determinants

Tells whether a matrix is invertible or not.

```python
np.linalg.det(A)
```

# 5. Full Linear Regression Using Linear Algebra

This is the real ML connection.

```python
w_best = np.linalg.inv(X.T @ X) @ X.T @ y
w_best
```

```python
pred_final = X @ w_best
pred_final
```

# 6. Compare Predictions

```python
pd.DataFrame({
    "Actual": y,
    "Predicted": pred_final
})
```

# Key Takeaways

- Matrix multiplication = ML prediction engine
- Transpose = required for formulas
- Inverse = solving equations (rare in practice)
- Determinant = checks matrix stability
- Linear regression = pure linear algebra
