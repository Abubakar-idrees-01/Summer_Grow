# 1. Vector Space

A vector space is a collection of all possible feature vectors.

In ML:
- Each row = a vector
- All rows together = vector space

This is called the feature space.

# 2. Linear Independence

Vectors are independent if none of them can be written using others.

In ML terms:
- Independent features = useful
- Dependent features = redundant

# 3. Basis

Basis = smallest set of independent vectors that describe the whole space.

In ML:
- Basis = most important features
- Everything else is derived from them

# 4. Span

Span = all possible combinations of vectors.

In ML:
- Span = all possible data representations from features

# 5. Feature Space Intuition

Each feature adds a new dimension:
- Area → 1D
- Area + Bedrooms → 2D
- Area + Bedrooms + Age → 3D
- All features → 4D space

# 6. Why this matters in Data Science

- ML models learn in vector space
- Each feature = a dimension
- Good features = independent directions
- Bad features = redundant directions

# 7. Practical Check: Redundant Feature Effect

```python
# Add duplicate feature
X_bad = np.column_stack([X, X[:,0]])

print("Original rank:", np.linalg.matrix_rank(X))
print("Bad rank:", np.linalg.matrix_rank(X_bad))
```

# Key Takeaways

- Vector space = all possible data points
- Linear independence = no redundant features
- Basis = minimal important features
- ML = learning structure in vector space
