# Linear Algebra for Data Science - Part 3
## Rank, Linear Systems, Gaussian Elimination

We continue using the same House Price dataset.

```python
import numpy as np
import pandas as pd

np.set_printoptions(suppress=True)

data = {
    'Area': [1000, 1500, 2000, 2500, 3000],
    'Bedrooms': [2, 3, 3, 4, 4],
    'Age': [10, 5, 20, 15, 8],
    'Distance': [5, 3, 10, 2, 7],
    'Price': [200000, 300000, 250000, 400000, 500000]
}

df = pd.DataFrame(data)

X = df[['Area','Bedrooms','Age','Distance']].to_numpy()
y = df['Price'].to_numpy()

X
```

# 1. Rank of a Matrix

Rank tells how many independent features exist in the dataset.

If rank is low → data has redundancy (bad for ML).

```python
np.linalg.matrix_rank(X)
```

# Why Rank Matters in Data Science

- Detect duplicate features
- Detect multicollinearity
- Improve regression stability
- Reduce overfitting risk

# 2. Linear System of Equations

We try to solve:

Xw = y

This is the core of Linear Regression.

```python
w = np.linalg.solve(X.T @ X, X.T @ y)
w
```

# 3. Gaussian Elimination (Concept)

Instead of inverse, we solve equations step-by-step.

This is what np.linalg.solve() does internally.

# 4. Rank vs ML Performance

If rank < number of features:
- Some features are useless
- Model becomes unstable
- Predictions become unreliable

```python
rank = np.linalg.matrix_rank(X)
features = X.shape[1]

rank, features
```

# 5. Detecting Redundant Features (Real ML Problem)

```python
# artificially add duplicate column
X_bad = np.column_stack([X, X[:,0]])

np.linalg.matrix_rank(X_bad)
```

Notice: Rank did not increase → duplicate feature detected.

# 6. Why ML cares about this

Rank problems lead to:
- Unstable regression coefficients
- Overfitting
- Poor generalization

# Key Takeaways

- Rank = number of independent features
- Low rank = redundant data
- Linear systems = foundation of regression
- np.linalg.solve() is better than inverse
- Bad rank = bad ML models
