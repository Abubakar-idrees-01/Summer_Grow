# Linear Algebra for Data Science - Part 6
## Eigenvalues, Eigenvectors, SVD, PCA

This is the core of modern Machine Learning and Data Science.

```python
import numpy as np
import pandas as pd

np.set_printoptions(suppress=True)

data = {
    'Area': [1000, 1500, 2000, 2500, 3000],
    'Bedrooms': [2, 3, 3, 4, 4],
    'Age': [10, 5, 20, 15, 8],
    'Distance': [5, 3, 10, 2, 7],
    'Price': [200000, 300000, 250000, 400000, 500000]
}

df = pd.DataFrame(data)
X = df[['Area','Bedrooms','Age','Distance']].to_numpy()
X
```

# 1. Eigenvalues & Eigenvectors

Eigenvectors are directions that do not change when a matrix is applied.

Eigenvalue = how much the vector is stretched.

```python
A = np.array([[2, 1],
              [1, 2]])

eigvals, eigvecs = np.linalg.eig(A)
eigvals, eigvecs
```

# Why it matters in Data Science

- PCA uses eigenvectors
- PageRank uses eigenvalues
- Clustering uses spectral methods
- Data variance directions come from eigenvectors

# 2. Covariance Matrix (Preparation for PCA)

Shows how features move together.

```python
X_centered = X - X.mean(axis=0)
cov_matrix = np.cov(X_centered.T)

cov_matrix
```

# 3. PCA (Principal Component Analysis)

PCA finds the most important directions (variance directions).

```python
eigvals, eigvecs = np.linalg.eig(cov_matrix)

eigvals
```

Sort eigenvectors by largest eigenvalues → principal components.

```python
idx = np.argsort(eigvals)[::-1]
components = eigvecs[:, idx]

components
```

# 4. Dimensionality Reduction

Project data onto fewer dimensions.

```python
X_pca = X_centered @ components[:, :2]
X_pca
```

# 5. Singular Value Decomposition (SVD)

SVD = more stable version of PCA.

```python
U, S, Vt = np.linalg.svd(X_centered)

S
```

# Why SVD matters

- Image compression
- Recommendation systems
- NLP embeddings (LSA)
- Noise reduction
- PCA implementation

# 6. Relationship: Eigen vs SVD

- Eigen → square matrices
- SVD → any matrix
- PCA uses both depending on implementation

# 7. Real Data Science Meaning

- Eigenvectors = important directions in data
- Eigenvalues = importance score
- PCA = remove noise, keep signal
- SVD = compress + simplify data

# Key Takeaways

- Eigenvectors = directions of maximum variance
- Eigenvalues = strength of those directions
- PCA = dimensionality reduction
- SVD = robust decomposition method
- These power modern ML systems
