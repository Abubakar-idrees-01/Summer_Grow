```python
import numpy as np
import pandas as pd

np.set_printoptions(suppress=True)
```

```python
# Dataset
data = {
    'Area': [1000, 1500, 2000, 2500, 3000],
    'Bedrooms': [2, 3, 3, 4, 4],
    'Age': [10, 5, 20, 15, 8],
    'Distance': [5, 3, 10, 2, 7],
    'Price': [200000, 300000, 250000, 400000, 500000]
}

df = pd.DataFrame(data)
df
```

# 1. Vectors

Each house = one vector (row in dataset)

```python
x1 = df.iloc[0][['Area','Bedrooms','Age','Distance']].to_numpy()
x2 = df.iloc[1][['Area','Bedrooms','Age','Distance']].to_numpy()

x1, x2
```

## Dot Product (Similarity between houses)

```python
np.dot(x1, x2)
```

# 2. Matrix Representation

Dataset = Matrix X

```python
X = df[['Area','Bedrooms','Age','Distance']].to_numpy()
y = df['Price'].to_numpy()

X, y
```

# 3. Matrix Multiplication (Core ML idea)

Prediction = X · weights

```python
w = np.array([100, 50000, -2000, -10000])  # fake weights

predictions = X @ w
predictions
```

# 4. Error (Difference between real and predicted)

```python
error = y - predictions
error
```

# 5. Transpose

Used in regression formulas

```python
X.T
```

# 6. Norm (Error size)

```python
np.linalg.norm(error)
```

# 7. Least Squares (Mini ML Model)

Find best weights using linear algebra

```python
w_best = np.linalg.inv(X.T @ X) @ X.T @ y
w_best
```

# 8. Final Prediction

```python
final_pred = X @ w_best
final_pred
```

# 9. Compare Real vs Predicted

```python
pd.DataFrame({
    'Actual': y,
    'Predicted': final_pred
})
```

# Key Takeaways

- Dataset = Matrix
- Each row = Vector
- Prediction = Matrix multiplication
- Learning = solving linear algebra equation
- Linear regression = pure linear algebra
