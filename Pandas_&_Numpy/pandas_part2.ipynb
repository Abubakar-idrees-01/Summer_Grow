```python
import pandas as pd

def clean_country(country):

    if pd.isna(country):
        return 'Unknown'

    country = str(country).strip().lower()

    # ---------- United States ----------
    us_keywords = [
        'united states',
        'united state',
        'u.s',
        'usa',
        'us of a',
        'the us',
        'the united states',
        'unitedstates'
    ]

    if any(k in country for k in us_keywords):
        return 'US'

    # ---------- United Kingdom ----------
    uk_keywords = [
        'united kingdom',
        'united kindom',
        'united kingdomk',
        'england',
        'wales',
        'northern ireland',
        'scotland',
        'u.k'
    ]

    if any(k in country for k in uk_keywords):
        return 'United Kingdom'

    # ---------- Canada ----------
    if 'canada' in country:
        return 'Canada'

    # ---------- Australia ----------
    if country.startswith('australi'):
        return 'Australia'

    # ---------- New Zealand ----------
    if 'new zealand' in country:
        return 'New Zealand'

    # ---------- Ireland ----------
    if country in ['ireland', 'republic of ireland']:
        return 'Ireland'

    # ---------- India ----------
    if 'india' == country:
        return 'India'

    # ---------- Germany ----------
    if 'germany' in country:
        return 'Germany'

    # ---------- Netherlands ----------
    if 'netherlands' in country:
        return 'Netherlands'

    # ---------- UAE ----------
    if 'united arab emirates' in country:
        return 'United Arab Emirates'

    # Default
    return country.title()
```

```python
df['country_clean'] = df['country'].apply(clean_country)
```
