# Exploratory Data Analysis — Wine Dataset (Seaborn, Part 2)

This notebook repeats the EDA from charts_wine.ipynb but uses Seaborn for visualization.
Adjust the `path` variable in the first code cell if your CSV is located elsewhere.

## 1 — Load dataset and quick overview
We load the CSV into a `DataFrame`, display the first rows and basic descriptive statistics.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_theme(style='whitegrid')
# Update this path if needed (relative to this notebook)
path = 'datasets/wine+quality/winequality-red-edited.csv'
df = pd.read_csv(path)

print('Shape:', df.shape)
display(df.head())
print('Descriptive statistics:')
display(df.describe())
```

## 2 — Missing values and basic checks
Check for missing data and duplicated rows.

```python
missing = df.isnull().sum()
print('Missing values per column:
', missing)
print('
Total duplicates:', df.duplicated().sum())
```

## 3 — Distribution of numerical features (histograms)
Use `sns.histplot` for cleaner histograms and KDE if desired.

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
n = len(num_cols)
cols_per_row = 3
rows = (n + cols_per_row - 1) // cols_per_row
fig, axes = plt.subplots(rows, cols_per_row, figsize=(5*cols_per_row, 4*rows))
axes = axes.flatten() if hasattr(axes, 'flatten') else [axes]
for i, col in enumerate(num_cols):
    sns.histplot(df[col].dropna(), bins=30, kde=False, color='tab:blue', ax=axes[i])
    axes[i].set_title(col)
# Hide unused subplots
for j in range(i+1, len(axes)):
    axes[j].axis('off')
fig.tight_layout()
plt.show()
```

## 4 — Boxplots to highlight outliers
Seaborn boxplots for each numeric feature.

```python
fig, ax = plt.subplots(figsize=(12,6))
sns.boxplot(data=df[num_cols], ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_title('Boxplots for numeric features')
plt.tight_layout()
plt.show()
```

## 5 — Quality counts (bar chart)
Use `sns.countplot` for counts per quality score.

```python
fig, ax = plt.subplots(figsize=(8,4))
sns.countplot(x='quality', data=df, palette='viridis', ax=ax)
ax.set_title('Distribution of Wine Quality')
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height + 1, int(height), ha='center')
plt.tight_layout()
plt.show()
```

## 6 — Scatter plots (example relationships)
Scatter plots using `sns.scatterplot`.

```python
fig, axes = plt.subplots(1, 2, figsize=(12,4))
sns.scatterplot(x='alcohol', y='quality', data=df, alpha=0.6, ax=axes[0], color='tab:purple')
axes[0].set_title('Alcohol vs Quality')
sns.scatterplot(x='volatile_acidity', y='quality', data=df, alpha=0.6, ax=axes[1], color='tab:orange')
axes[1].set_title('Volatile Acidity vs Quality')
plt.tight_layout()
plt.show()
```

## 7 — Correlation matrix (heatmap with Seaborn)
Use `sns.heatmap` to show Pearson correlations with annotations.

```python
corr = df[num_cols].corr()
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu', vmin=-1, vmax=1, ax=ax)
ax.set_title('Correlation matrix (Pearson)')
plt.tight_layout()
plt.show()
```
